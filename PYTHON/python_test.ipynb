```python
def in_cache(func):
    cache = {}
    def wrapper(n):
        if n in cache:
            return cache[n]
        else:
            cache[n] = func(n)
            return cache[n]
    return wrapper
```

```python
def factorial(n):
    ret = 1
    for i in range(1,n+1):
        ret *= i
    return ret
```

```python
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter
```

```python
counter_a = make_counter()
```
